```python
import cv2
fmt = r'C:\Users\cidzerda\Documents\GitHub\streams\{}.mp4'
file_titles = (
    'anytimeshield',
    'nikolarn',
    'shivfps',
    'staycationyoutube',
    'tannerslays',
    'thefyzu',
)
window_title = 'Find Legend Crop'
def open_file(index):
    c = cv2.VideoCapture(fmt.format(file_titles[index]), cv2.CAP_MSMF)
    succeeded, frame = c.read()
    if succeeded:
        cv2.imshow(window_title, frame)
    return c, frame
index = 0
c, frame = open_file(index)
top = left = 0
bottom = len(frame)
right = len(frame[0])
```

```python
while True:
    part = frame[top:bottom, left:right]
    if len(part[0]) < 200:
        part = cv2.resize(part, (200, 200 * len(part) // len(part[0])))
    cv2.imshow('Crop', part)
    key = chr(cv2.waitKey(0))
    if key == 'o':
        c.release()
        g = ('{}\t{}'.format(*t) for t in enumerate(file_titles))
        print(*g, sep='\n')
        index = int(input('Enter file index: '))
        c, frame = open_file(index)
    elif key == '?':
        fmt = '({}, {}) -> ({}, {}) is height {}, width {}'
        print(fmt.format(top, left, bottom, right, bottom - top, right - left))
    elif key == 'p':
        c.release()
        index = max(0, index - 1)
        c, frame = open_file(index)
    elif key == 'n':
        c.release()
        index = min(index + 1, len(file_titles) - 1)
        c, frame = open_file(index)
    elif key == 'c':
        top, left, bottom, right = map(int, input('top, left, bottom, right: ').split())
    elif key == '#':
        c.set(cv2.CAP_PROP_POS_FRAMES, int(input('Enter frame index: ')))
        _, frame = c.read()
        cv2.imshow(window_title, frame)
    elif key == ',':
        c.set(cv2.CAP_PROP_POS_FRAMES, max(0, c.get(cv2.CAP_PROP_POS_FRAMES) - 2))
        _, frame = c.read()
        cv2.imshow(window_title, frame)
    elif key == '.':
        _, frame = c.read()
        cv2.imshow(window_title, frame)
    elif key == '<':
        c.set(cv2.CAP_PROP_POS_FRAMES, max(0, c.get(cv2.CAP_PROP_POS_FRAMES) - 11))
        _, frame = c.read()
        cv2.imshow(window_title, frame)
    elif key == '>':
        c.set(cv2.CAP_PROP_POS_FRAMES, c.get(cv2.CAP_PROP_POS_FRAMES) + 9)
        _, frame = c.read()
        cv2.imshow(window_title, frame)
    elif key == 'e':
        top = max(0, top - 1)
    elif key == 's':
        left = max(0, left - 1)
    elif key == 'd':
        top = min(top + 1, bottom - 1)
    elif key == 'f':
        left = min(left + 1, right - 1)
    elif key == 'E':
        top //= 2
    elif key == 'S':
        left //= 2
    elif key == 'D':
        top += (bottom - top) // 2
    elif key == 'F':
        left += (right - left) // 2
    elif key == 'i':
        bottom = max(top + 1, bottom - 1)
    elif key == 'j':
        right = max(left + 1, right - 1)
    elif key == 'k':
        bottom = min(bottom + 1, len(frame))
    elif key == 'l':
        right = min(right + 1, len(frame[0]))
    elif key == 'I':
        bottom -= (bottom - top) // 2
    elif key == 'J':
        right -= (right - left) // 2
    elif key == 'K':
        bottom += (len(frame) - bottom) // 2
    elif key == 'L':
        right += (len(frame[0]) - right) // 2
    else:
        break
```

```python
c.release()
cv2.destroyAllWindows()
```
