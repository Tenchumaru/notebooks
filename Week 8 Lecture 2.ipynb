Assume this schema is used in an ecommerce site. Provide Python code that generates random user interactions for use in a machine learning training session.

Below is a practical Python example that simulates **ecommerce user–item interactions** suitable for ML training (e.g., recommendation systems, conversion prediction, churn modeling).

The code generates **behavioral events** such as views, add-to-cart, and purchases with realistic probabilities.

**Interaction Schema (Generated)**

Each interaction includes:

- `user_id`
- `item_id`
- `event_type` (view, add_to_cart, purchase)
- `timestamp`
- `price`

```python
import random
import uuid
from datetime import datetime, timedelta
import pandas as pd

# Configuration
NUM_USERS = 200
NUM_ITEMS = 100
NUM_INTERACTIONS = 2000

COUNTRIES = ["US", "CA", "UK", "IN"]
CATEGORIES = ["Electronics", "Home", "Beauty", "Clothing", "Books"]
EVENT_TYPES = ["view", "add_to_cart", "purchase"]

EVENT_PROBABILITIES = {
    "view": 0.70,
    "add_to_cart": 0.20,
    "purchase": 0.10
}

# Generate users
users = [
    {
        "user_id": str(uuid.uuid4()),
        "age": random.randint(18, 75),
        "is_premium": random.choice([True, False]),
        "country_code": random.choice(COUNTRIES),
    }
    for _ in range(NUM_USERS)
]

# Generate items
items = [
    {
        "item_id": str(uuid.uuid4()),
        "category": random.choice(CATEGORIES),
        "price": round(random.uniform(5.0, 500.0), 2),
    }
    for _ in range(NUM_ITEMS)
]

def random_timestamp(days_back=60):
    return datetime.now() - timedelta(
        days=random.randint(0, days_back),
        seconds=random.randint(0, 86400),
    )

# Generate interactions
interactions = []

for _ in range(NUM_INTERACTIONS):
    user = random.choice(users)
    item = random.choice(items)

    event_type = random.choices(
        population=EVENT_TYPES,
        weights=[
            EVENT_PROBABILITIES["view"],
            EVENT_PROBABILITIES["add_to_cart"],
            EVENT_PROBABILITIES["purchase"],
        ],
    )[0]

    interaction = {
        "user_id": user["user_id"],
        "item_id": item["item_id"],
        "event_type": event_type,
        "timestamp": random_timestamp(),
        "price": item["price"],
        "category": item["category"],
        "country_code": user["country_code"],
        "is_premium": user["is_premium"],
        # Supervised learning label
        "label_purchase": 1 if event_type == "purchase" else 0,
    }

    interactions.append(interaction)

# Convert to DataFrame
df = pd.DataFrame(interactions)

# Sort chronologically (important for sequence models)
df = df.sort_values("timestamp").reset_index(drop=True)

print(df.head())
```

Below is a **behavior-aware data generator** specifically designed for **predicting user behavior** (e.g., _will the user purchase next?_, _will they add to cart?_).
It produces **sequential interactions**, **realistic funnels**, and **supervised labels** suitable for most ML models.

---

**Modeling Assumptions**

- Users have latent preferences for categories
- Behavior follows a funnel: view → add_to_cart → purchase
- Premium users convert more often
- Past behavior influences future actions
- Output supports **classification, ranking, and sequence models**

---

**Output Schema (Training-Ready)**

Each row represents **one interaction**:

- `user_id`
- `item_id`
- `event_type` (view, add_to_cart, purchase)
- `timestamp`
- `category`
- `price`
- `country_code`
- `is_premium`
- `user_category_affinity`
- `label_next_purchase` ← **prediction target**

```python
import random
import uuid
from datetime import datetime, timedelta
import pandas as pd

# -----------------------
# Configuration
# -----------------------
NUM_USERS = 300
NUM_ITEMS = 200
MAX_EVENTS_PER_USER = 30

CATEGORIES = ["Electronics", "Home", "Beauty", "Clothing", "Books"]
COUNTRIES = ["US", "CA", "UK", "IN"]

# Base probabilities
VIEW_TO_CART = 0.25
CART_TO_PURCHASE = 0.30
PREMIUM_BOOST = 1.4

# -----------------------
# Users with preferences
# -----------------------
users = []
for _ in range(NUM_USERS):
    preferred_category = random.choice(CATEGORIES)
    users.append({
        "user_id": str(uuid.uuid4()),
        "country_code": random.choice(COUNTRIES),
        "is_premium": random.choice([True, False]),
        "preferred_category": preferred_category
    })

# -----------------------
# Items
# -----------------------
items = []
for _ in range(NUM_ITEMS):
    items.append({
        "item_id": str(uuid.uuid4()),
        "category": random.choice(CATEGORIES),
        "price": round(random.uniform(10, 600), 2)
    })

# -----------------------
# Helper functions
# -----------------------
def category_affinity(user, item):
    return 1.0 if user["preferred_category"] == item["category"] else 0.3

def event_time(base_time, step):
    return base_time + timedelta(minutes=step * random.randint(5, 30))

# -----------------------
# Generate interactions
# -----------------------
rows = []
base_time = datetime.now() - timedelta(days=60)

for user in users:
    num_events = random.randint(10, MAX_EVENTS_PER_USER)
    current_time = base_time

    for i in range(num_events):
        item = random.choice(items)
        affinity = category_affinity(user, item)

        # View event
        rows.append({
            "user_id": user["user_id"],
            "item_id": item["item_id"],
            "event_type": "view",
            "timestamp": current_time,
            "category": item["category"],
            "price": item["price"],
            "country_code": user["country_code"],
            "is_premium": user["is_premium"],
            "user_category_affinity": affinity,
            "label_next_purchase": 0
        })

        # Add to cart
        cart_prob = VIEW_TO_CART * affinity
        if user["is_premium"]:
            cart_prob *= PREMIUM_BOOST

        if random.random() < cart_prob:
            current_time = event_time(current_time, 1)
            rows.append({
                **rows[-1],
                "event_type": "add_to_cart",
                "timestamp": current_time
            })

            # Purchase
            purchase_prob = CART_TO_PURCHASE * affinity
            if user["is_premium"]:
                purchase_prob *= PREMIUM_BOOST

            if random.random() < purchase_prob:
                current_time = event_time(current_time, 1)
                rows.append({
                    **rows[-1],
                    "event_type": "purchase",
                    "timestamp": current_time,
                    "label_next_purchase": 1
                })

        current_time = event_time(current_time, 1)

# -----------------------
# Final dataset
# -----------------------
df = pd.DataFrame(rows).sort_values(
    ["user_id", "timestamp"]
).reset_index(drop=True)

print(df.head())
```

**Shared Preprocessing Pipeline**

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, roc_auc_score
```

```python
# Target and features
TARGET = "label_next_purchase"

NUM_FEATURES = [
    "price",
    "user_category_affinity",
]

CAT_FEATURES = [
    "event_type",
    "category",
    "country_code",
    "is_premium",
]

X = df[NUM_FEATURES + CAT_FEATURES]
y = df[TARGET]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# Preprocessing
numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("onehot", OneHotEncoder(
        handle_unknown="ignore",
        sparse_output=False
    ))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, NUM_FEATURES),
        ("cat", categorical_transformer, CAT_FEATURES),
    ]
)
```

1️⃣ **Logistic Regression Pipeline**

```python
from sklearn.linear_model import LogisticRegression

logreg_pipeline = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        n_jobs=-1
    ))
])

logreg_pipeline.fit(X_train, y_train)

y_pred = logreg_pipeline.predict(X_test)
y_proba = logreg_pipeline.predict_proba(X_test)[:, 1]

print("Logistic Regression Results")
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
```

**Why Logistic Regression?**

- Strong baseline
- Fast training
- Probabilistic output
- Feature interpretability

---

2️⃣ **Random Forest Pipeline**

```python
from sklearn.ensemble import RandomForestClassifier

rf_pipeline = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", RandomForestClassifier(
        n_estimators=200,
        max_depth=12,
        min_samples_leaf=5,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])

rf_pipeline.fit(X_train, y_train)

y_pred = rf_pipeline.predict(X_test)
y_proba = rf_pipeline.predict_proba(X_test)[:, 1]

print("Random Forest Results")
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
```

**Why Random Forest?**

- Captures non-linear behavior
- Handles feature interactions well
- Robust to noise
- Minimal feature engineering needed
