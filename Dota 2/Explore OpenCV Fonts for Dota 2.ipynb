```python
import cv2
import itertools as it
import numpy as np
import random

from utilities import *
```

```python
width, height = 200, 14
alphabet_encoding = [chr(i) for i in range(ord('A'), ord('Z') + 1)] + list("'-")
alphabet_dict = {alphabet_encoding[i]: i for i in range(len(alphabet_encoding))}

min_text_length, max_text_length = 2, 19

population = alphabet_encoding + len(alphabet_encoding) // 5 * [' ']
```

```python
def make_frame_generator(file_title):
    video = cv2.VideoCapture(fr"C:\Users\cidzerda\Documents\GitHub\strevr-dota-2\heroes\{file_title}.mp4")
    for i in range(9):
        video.set(cv2.CAP_PROP_POS_FRAMES, i * 111)
        _, frame = video.read()
        yield frame[609:623, 348:548, :]
    video.release()
```

```python
# Determine the appropriate font scales to use for each OpenCV font.

font_names = [s for s in dir(cv2) if s.startswith('FONT_HERSHEY_') and not s.startswith('FONT_HERSHEY_SCRIPT_')]
font_scales = [0.321, 0.452, 0.336, 0.676, 0.341, 0.318] # These are best.
font_scales = [0.33, 0.47, 0.35, 0.71, 0.35, 0.33] # These are slightly larger than above.

frame = np.vstack(list(make_frame_generator('queen_of_pain')))

def fn(i):
    # Construct a random background.
    font_name = font_names[i]
    print(font_name)
    color = (256 * np.random.rand(3)).astype(np.uint8)
    font_scale = font_scales[i]
    text = 'QUEEN OF PAIN'
    x = 57
    origin, font_face, line_type = (x, 10), getattr(cv2, font_name), cv2.LINE_AA
    while True:
        image = color.reshape((1, 1, 3))
        image = cv2.resize(image, (width, height))

        # Put some text in it.
        cv2.putText(image, text, origin, font_face, font_scale, color=[int(c / 4) for c in color], thickness=2, lineType=line_type)
        r, g, b = (220, 225, 255)
        cv2.putText(image, text, origin, font_face, font_scale, color=(b, g, r), thickness=1, lineType=line_type)

        ch = show_and_wait(np.vstack([frame, image, frame]))
        if ch == '+':
            font_scale *= 1.003
        elif ch == '-':
            font_scale /= 1.003
        else:
            break

    return image, font_scale

l = [fn(i) for i in range(len(font_names))]
print(*(font_scale for _, font_scale in l), sep=', ')
show_and_wait(np.vstack([frame] + [image for image, _ in l] + [frame]))
cv2.destroyAllWindows()
```

```python
# Test the font scale ranges and maximum horizontal offsets for the selected OpenCV fonts.

font_names = ('FONT_HERSHEY_DUPLEX', 'FONT_HERSHEY_PLAIN', 'FONT_HERSHEY_SIMPLEX')
font_scale_ranges = ((0.27, 0.35), (0.55, 0.71), (0.28, 0.35))
fonts = tuple(getattr(cv2, s) for s in font_names)
maximum_horizontal_offsets = (30, 175)

def make_random_text():
    while True:
        n = random.randint(min_text_length, max_text_length)
        l = random.choices(population, k=n)
        image_text = ''.join(l)
        label_text = image_text.replace(' ', '')
        if len(label_text) >= min_text_length:
            return image_text.strip(), label_text

def make_pair(transform=lambda a, b: (a, b)):
    # Construct a random background.
    color = (256 * np.random.rand(3)).astype(np.uint8)
    image = color.reshape((1, 1, 3))
    image = cv2.resize(image, (width, height))

    # Put some text in it.
    image_text, label_text = make_random_text()
    x = random.randint(5, 30 + int((175 - 30) * (max_text_length - len(image_text)) / (max_text_length - min_text_length)))
    font_index = random.randrange(0, len(fonts))
    origin, font_face, line_type = (x, 10), fonts[font_index], cv2.LINE_AA
    font_scale = random.uniform(*font_scale_ranges[font_index])
    color = [int(c / 4) for c in color]
    cv2.putText(image, image_text, origin, font_face, font_scale, color=color, thickness=2, lineType=line_type)
    r, g, b = (210, 185, 141) if random.random() < .5 else (220, 225, 255)
    cv2.putText(image, image_text, origin, font_face, font_scale, color=(b, g, r), thickness=1, lineType=line_type)
    if not b:
        cv2.imshow('tesst', image)
        cv2.waitKey()
        cv2.destroyAllWindows()

    return transform(image, label_text)

l = [make_pair() for _ in range(88)]
print(*(text for _, text in l), sep=',')
show_and_wait(np.vstack([image for image, _ in l]))
cv2.destroyAllWindows()
```
