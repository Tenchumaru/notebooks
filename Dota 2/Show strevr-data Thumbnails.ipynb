```python
import json

with open(r"C:\Users\Chris\Documents\GitHub\strevr-data\Dota 2\Heroes\parameters.json") as fin:
    parameters = json.load(fin)
d = parameters['cropping_bounds']
g = ('top', 'left', 'bottom', 'right')
g = (d[k] for k in g)
top, left, bottom, right = g
top, left, bottom, right
```

```python
import cv2
import numpy as np
import sys
from pathlib import Path
from utilities import create_square_image, show_and_wait

def fn(path):
    video = cv2.VideoCapture(str(path), cv2.CAP_MSMF if sys.platform == 'win32' else cv2.CAP_ANY)
    result, frame = video.read()
    return frame[top:bottom, left:right] if result else None

path = Path(r'C:\Users\Chris\Documents\GitHub\strevr-data\Dota 2\Heroes')
g = (fn(p) for p in path.glob('**/0000000.mpeg'))
images = [i for i in g if i is not None]
```

```python
image = create_square_image(images)
```
